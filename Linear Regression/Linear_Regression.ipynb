```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import statsmodels.api as sm
import pandas as pd
from sklearn.linear_model import LinearRegression
#Initialising the libraries and data points for calculating regression

area = np.array([1,2,3,7,8,9,10,12,15,17,21])
prices = np.array([150, 250,330, 460, 510, 625,690, 780, 754, 880, 950])

plt.scatter(area, prices)
plt.xlabel("area")
plt.ylabel("prices")
```

So, now we need to update our existing 'm' and 'c' values in this way:

m(t+1) = f(m(t), error(t))


c(t+1) = f(c(t), error(t))

```python

def cost(iterations, L, x, y):
  #This function returns the average error between the given values and predicted values.
  m = 0
  c = 0
  L = 0.001
  n = float(len(x))
  for i in range(iterations):

    y_predicted = m*x + c
    D_m = (-2/n)*sum(x*(y-y_predicted))
    D_c = (-2/n)*sum(y-y_predicted)
    m = m - L*D_m
    c = c-L*D_c
    if i%2==0 and i<50:
        print("Epoch {}".format(i))
        plt.scatter(x,y)
        plt.plot([min(x), max(x)], [min(y_predicted), max(y_predicted)], color='red')  # regression line
        plt.show()

  print (m, c)
  return m,c

m, c = cost(1000, 0.001, area, prices)

def prediction(x, m, c):
  return (m*x +c)
```

```python
area = 19
print(prediction(area, m, c))
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
print(x)
y = 2 * x - 8 + rng.randn(50)
plt.scatter(x, y);
```

```python
model = LinearRegression(fit_intercept=True)
#We utilise the Linear Regression function from Scikit learn so that we dont have to write the function everytime
model.fit(x[:, np.newaxis], y)

xfit = np.linspace(min(x), max(x), 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit, color = "red");
```
